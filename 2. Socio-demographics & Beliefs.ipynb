```python
#Importing all libraries needed
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
import os
# Set Pandas to display all columns
pd.set_option('display.max_columns', None)
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
%matplotlib inline
```

```python
# Ensure the folder exists
output_folder = "Figures/ColumnsAnalysis/"
os.makedirs(output_folder, exist_ok=True)
```

```python
# Define a color palette for initial endowment
wealth_palette = {
    "24": "#aec7e8",  # Light Blue
    "30": "#ffbb78",  # Light Orange
    "40": "#98df8a",  # Light Green
    "48": "#ff9896",  # Light Red
    "60": "#c5b0d5",  # Light Purple
}
```

```python
#Reading the users dataset
df_users = pd.read_csv("Datasets/df_users_six_players.csv")
print("Gender", df_users.gender.unique())
print("Age range", df_users.age_range.unique())
print("Educational level", df_users.educational_level.unique())
print("Economic status", df_users.economic_status.unique())
print("Working status", df_users.working_status.unique())
print("Residence", df_users.residence.unique())
```

```python
# Define mappings for better labels
label_mappings = {
    "residence": {
        "r1": "Ciutat Vella", "r2": "L'Eixample", "r3": "Sants-Montjuïc", "r4": "Les Corts",
        "r5": "Sarrià - Sant Gervasi", "r6": "Gràcia", "r7": "Horta-Guinardó", "r8": "Nou Barris",
        "r9": "Sant Andreu", "r10": "Sant Martí", "r11": "Out of Barcelona"
    },
    "gender": {"M": "Male", "F": "Female", "O": "Other", "N": "No answer"},
    "age_range": {
        "r1": "14-19", "r2": "20-24", "r3": "25-34", "r4": "35-44",
        "r5": "45-54", "r6": "55-64", "r7": "65+"
    },
    "educational_level": {
        "r1": "None", "r2": "Primary", "r3": "Secondary",
        "r4": "Higher", "r5": "No answer"
    },
    "economic_status": {
        "r1": "No Income", "r2": "Low income", "r3": "Middle income",
        "r4": "High income", "r5": "No answer"
    },
    "working_status" : {
        "r1": "None", "r2": "Studying", "r3": "Working",
        "r4": "Studying & Working", "r5": "No answer" 
    }  
}

# Convert all categorical columns to string type before mapping
for col in label_mappings.keys():
    df_users[col] = df_users[col].astype(str)  # Ensure all values are strings
    df_users[col] = df_users[col].map(label_mappings[col])  # Apply label mappings

# Drop any rows where 'endowment_initial' is missing
df_users = df_users.dropna(subset=['endowment_initial'])
print("Gender", df_users.gender.unique())
print("Age range", df_users.age_range.unique())
print("Educational level", df_users.educational_level.unique())
print("Economic status", df_users.economic_status.unique())
print("Working status", df_users.working_status.unique())
print("Residence", df_users.residence.unique())
```

Number of participants across socio-demographic characteristics and initial endowments

```python
# Define user characteristic columns
user_characteristics = ['gender', 'age_range', 'educational_level', 'economic_status','working_status', 'residence']
```

```python
#STACKED BAR PLOTS BY INITIAL ENDOWMENT
for col in user_characteristics:
    # Compute counts with 'endowment_initial' as string
    crosstab = pd.crosstab(df_users[col], df_users['endowment_initial'])

    # Compute total counts for sorting
    total_counts = crosstab.sum(axis=1).sort_values(ascending=False)

    # Reorder rows based on total count (highest to lowest)
    crosstab = crosstab.loc[total_counts.index]

    # Ensure column names are strings for proper color mapping
    crosstab.columns = crosstab.columns.astype(str)

    # Create figure
    fig, ax = plt.subplots(figsize=(8.5, 6))

    # Create stacked bars with pastel colors, broad width, and black contours
    crosstab.plot(kind='bar', stacked=True, ax=ax, width=0.9,
                  color=[wealth_palette[k] for k in crosstab.columns], edgecolor="black", linewidth=0.8, legend=False)

    # Labels & Formatting
    plt.xlabel("")
    plt.ylabel("Number of players", fontsize=18)
    
    if col == "gender":
        plt.xticks(rotation=0, ha='center', fontsize=15)
        plt.xlabel("Gender (ranked)", fontsize=18)

    if col == "age_range":
        plt.xticks(rotation=0, ha='center', fontsize=15)
        plt.xlabel("Age range (ranked)", fontsize=18)

    if col == "educational_level":
        plt.xticks(rotation=0, ha='center', fontsize=15)
        plt.xlabel("Level of education (ranked)", fontsize=18)

    if col == "economic_status":
        plt.xticks(rotation=0, ha='center', fontsize=15)
        plt.xlabel("Economic status (ranked)", fontsize=18)

    if col == "working_status":
        plt.xticks(rotation=25, ha='right', fontsize=15)
        plt.xlabel("Working status (ranked)", fontsize=18)

    if col == "residence" or col =="":
        plt.xticks(rotation=45, ha='right', fontsize=15)
        plt.xlabel("Neighbourhood of residence (ranked)", fontsize=18)
    
    # Completely remove legend if created
    if ax.get_legend() is not None:
        ax.get_legend().remove()

    plt.tick_params(labelcolor='black', labelsize=16)

    # Save
    plt.savefig(f"{output_folder}/Stacked_{col}_Distribution_Histogram.pdf", bbox_inches='tight', pad_inches=0.1)
    plt.show()
```

```python
#NORMALIZED STACKED BAR PLOTS BY INITIAL ENDOWMENT (0 to 1 SCALE)
for col in user_characteristics:
    # Compute counts with 'endowment_initial' as string
    crosstab = pd.crosstab(df_users[col], df_users['endowment_initial'])

    # Compute total counts for sorting
    total_counts = crosstab.sum(axis=1).sort_values(ascending=False)

    # Reorder rows based on total count (highest to lowest)
    crosstab = crosstab.loc[total_counts.index]

    # Normalize row-wise (each row sums to 1)
    crosstab_norm = crosstab.div(crosstab.sum(axis=1), axis=0)

    # Ensure column names are strings for proper color mapping
    crosstab_norm.columns = crosstab_norm.columns.astype(str)

    # Create figure
    fig, ax = plt.subplots(figsize=(8.5, 6))

    # Create stacked bars with pastel colors, broad width, and black contours
    crosstab_norm.plot(kind='bar', stacked=True, ax=ax, width=0.9,
                       color=[wealth_palette[k] for k in crosstab_norm.columns], edgecolor="black", linewidth=0.8)

    # Labels & Formatting
    plt.xlabel("")
    plt.ylabel("Proportion of players", fontsize=18)
    
    user_characteristics = ['gender', 'age_range', 'educational_level', 'economic_status','working_status', 'residence']
    if col == "gender":
        plt.xticks(rotation=0, ha='center', fontsize=15)
        plt.xlabel("Gender (ranked)", fontsize=18)

    if col == "age_range":
        plt.xticks(rotation=0, ha='center', fontsize=15)
        plt.xlabel("Age range (ranked)", fontsize=18)

    if col == "educational_level":
        plt.xticks(rotation=0, ha='center', fontsize=15)
        plt.xlabel("Level of education (ranked)", fontsize=18)

    if col == "economic_status":
        plt.xticks(rotation=0, ha='center', fontsize=15)
        plt.xlabel("Economic status (ranked)", fontsize=18)

    if col == "working_status":
        plt.xticks(rotation=25, ha='right', fontsize=15)
        plt.xlabel("Working status (ranked)", fontsize=18)

    if col == "residence" or col =="":
        plt.xticks(rotation=45, ha='right', fontsize=15)
        plt.xlabel("Neighbourhood of residence (ranked)", fontsize=18)
    
    # Completely remove legend if created
    if ax.get_legend() is not None:
        ax.get_legend().remove()

    plt.tick_params(labelcolor='black', labelsize=16)

    # Save
    plt.savefig(f"{output_folder}/Normalized_Stacked_{col}_Distribution.pdf", bbox_inches='tight', pad_inches=0.1)
    plt.show()
```

Distribution of participant’ responses to belief and attitudinal questions

```python
# Define belief characteristics to analyze
beliefs_characteristics = ["frame_pr1", "frame_pr2", "frame_pr3", "enquesta_final_pr1",
                           "enquesta_final_pr2", "enquesta_final_pr3", "enquesta_final_pr4",
                           "enquesta_final_pr5", "enquesta_final_pr6", "enquesta_final_pr7",
                           "enquesta_final_pr8", "enquesta_final_pr9", "enquesta_final_pr10"]
                           # Define belief characteristics for labeling
beliefs_characteristics1 = ["frame_pr1", "frame_pr2", "frame_pr3", "final_pr1",
                           "final_pr2", "final_pr3", "final_pr4",
                           "final_pr5", "final_pr6", "final_pr7",
                           "final_pr8", "final_pr9", "final_pr10"]
```

```python
# Initialize figure
fig, ax = plt.subplots(figsize=(12, 6))

# Dictionary to store crosstab results for stacking
stacked_data = {}

# Compute stacked data for each category
for col in beliefs_characteristics:
    crosstab = df_users[col].value_counts().sort_index()  # Get counts
    stacked_data[col] = crosstab

# Set bar positions
bar_width = 0.7  # Bar width
bar_positions = np.arange(len(beliefs_characteristics))  # Position of bars
bottoms = np.zeros(len(beliefs_characteristics))  # Bottom positions for stacking

# Define a color map
color_palette = plt.get_cmap("Set2")

# Iterate through each category and plot stacked bars
for col_idx, col in enumerate(beliefs_characteristics):
    crosstab = stacked_data[col]

    for idx, option in enumerate(crosstab.index):
        counts = crosstab[option] if option in crosstab.index else 0
        ax.bar(bar_positions[col_idx], counts, width=bar_width,
               color=color_palette(idx % 10), edgecolor="black", linewidth=0.8,
               bottom=bottoms[col_idx], label=option if col_idx == 0 else "")

        # Update bottom for stacking
        bottoms[col_idx] += counts

# Formatting
ax.set_xticks(bar_positions)
ax.set_xticklabels(beliefs_characteristics1, rotation=45, ha="right", fontsize=18)
ax.set_ylabel("Number of players", fontsize=18)
ax.tick_params(labelcolor='black', labelsize=18)
all_options=["r1", "r2", "r3", "r4", "r5"]
# Define a color map (dictionary) for each option
all_options = ["r1", "r2", "r3", "r4", "r5"]
color_map = {opt: color_palette(i) for i, opt in enumerate(all_options)}

handles = [plt.Rectangle((0,0),1,1, color=color_map[opt]) for opt in all_options]
ax.legend(handles, all_options, title="Response codes", loc='lower left', fontsize=12, frameon=True)

# Show plot
plt.tight_layout()
# Save
plt.savefig(f"{output_folder}/Overall_distribution_Beliefs_Charactertistics.pdf", bbox_inches='tight', pad_inches=0.1)
plt.show()
```
