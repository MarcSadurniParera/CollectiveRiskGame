```python
# Load Synthetic Simulations for All Hypothesis
simulated_dfs = {
    "H0": pd.read_csv("Datasets/df_users_H0.csv"),
    "H1": pd.read_csv("Datasets/df_users_H1.csv"),
    "H2": pd.read_csv("Datasets/df_users_H2.csv"),
    "H3": pd.read_csv("Datasets/df_users_H3.csv"),
    "Hybridisation": pd.read_csv("Datasets/df_users_Hybridisation.csv")
}
# Load Experimental datasets for comparision
df_data_macro = pd.read_csv("Datasets/Observables_Data.csv")
successful_users_six = pd.read_csv("Datasets/successful_users_six.csv")
```

Filter Successful Users and computing Normalised payoff

```python
simulated_dfs_success = {}

for hypo, df in simulated_dfs.items():
    successful = df[df["goal_reached"] == 1].copy()  # <- add .copy() here
    # Compute normalized payoffs for synthetic data
    successful['normalized_payoff'] = successful['endowment_current'] / (0.5 * successful['endowment_initial'])
    simulated_dfs_success[hypo] = successful

# Compute normalized payoffs for real data
successful_users_six['normalized_payoff'] = successful_users_six['endowment_current'] / (0.5 * successful_users_six['endowment_initial'])
```
